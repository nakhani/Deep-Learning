```python
import os
import pandas as pd
from deepface import DeepFace

def generate_combined_dataset(input_folder, output_file, model_name="ArcFace"):


    data = []


    for label in os.listdir(input_folder):
        label_path = os.path.join(input_folder, label)
        if os.path.isdir(label_path):
            for file in os.listdir(label_path):
                file_path = os.path.join(label_path, file)
                if file.endswith(('.jpg', '.png', '.jpeg')):
                    try:

                        embedding_objs = DeepFace.represent(img_path=file_path, model_name=model_name)
                        features = embedding_objs[0]['embedding']


                        row = [label, file] + features
                        data.append(row)
                    except Exception as e:
                        print(f"Error processing {file_path}: {e}")


    if data:
        columns = ['label', 'file_name'] + [f'feature{i+1}' for i in range(len(data[0]) - 2)]
        dataset = pd.DataFrame(data, columns=columns)


        dataset.to_csv(output_file, index=False)
        print(f"Combined dataset saved to {output_file}")
    else:
        print(f"No valid images found in the directory: {input_folder}")


input_folder = "/content/drive/MyDrive/dataset/persian_dataset"
output_file = "combined_dataset.csv"
generate_combined_dataset(input_folder, output_file, model_name="ArcFace")
```
