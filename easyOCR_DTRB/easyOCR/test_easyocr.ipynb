```python
import easyocr
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
```

```python
image = mpimg.imread("Handwriting.jpg")

reader = easyocr.Reader(['en'])  
result = reader.readtext('Handwriting.jpg')

plt.imshow(image)
plt.axis("off")  
plt.show()
print(result)
```

```python
image = mpimg.imread("persian_handwriting.png")

reader = easyocr.Reader(['fa'])  
result = reader.readtext('persian_handwriting.png')

plt.imshow(image)
plt.axis("off")  
plt.show()
print(result)
```
